```python
#Exercise Day 3
#L5_1 Image Subtraction
#Student: Nguyen Thi Minh Hien

#Import libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os

#read image
liveImg = cv2.imread('/content/Images/live.jpg');
maskImg = cv2.imread('/content/Images/mask.jpg');

#calculate the difference between two images
diffImg = np.abs(cv2.subtract(maskImg, liveImg));
gamma = 0.5;
diffImg_contrasted = diffImg ** gamma;

#Create a subplot with rows, columns to show multiple image
fig, ax = plt.subplots(2,2, figsize=(12,14));

# show each image on subplot
ax[0,0].imshow(liveImg);
ax[0,0].set_title('Original Live image')

ax[0,1].imshow(maskImg);
ax[0,1].set_title('Original Mask image')

ax[1,0].imshow(diffImg);
ax[1,0].set_title('Difference Image');

ax[1,1].imshow((diffImg_contrasted*255).astype(np.uint8));
ax[1,1].set_title('Difference Image increased contrast')

#Show plot super title
plt.suptitle('L5_1 Image Subtraction', fontsize=18);
plt.show();

# to specified directory 
directory = '/content/Images';
os.chdir(directory);
#Save images
cv2.imwrite('Image_Subtraction.jpg', diffImg);
cv2.imwrite('Image_Subtraction_with_contrast.jpg', diffImg_contrasted);
```
